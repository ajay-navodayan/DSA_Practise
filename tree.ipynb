Binary Tree preorder

```python
class Solution {
    public List<Integer> preorderTraversal(TreeNode root) {
        ArrayList<Integer> Arlist=new ArrayList<>();
        helper(root,Arlist);
        return Arlist;
        
    }
    public static void helper(TreeNode root, ArrayList<Integer> Arlist){
        if(root==null){
            return;
        }
        Arlist.add(root.val);
        helper(root.left,Arlist);
        helper(root.right,Arlist);
        return;
    }
}
```

Binary Tree Inorder Traversal


```python
class Solution {
    public List<Integer> inorderTraversal(TreeNode root) {
        ArrayList<Integer> ar=new ArrayList<>();
        helper(root,ar);
        return ar;
        
    }
    private static void helper(TreeNode root, ArrayList<Integer> ar){
        if(root==null){
            return;
        }
        helper(root.left, ar);
        ar.add(root.val);
        helper(root.right,ar);
        return;
    }
}
```

Binary Tree Postorder Traversal

```python
class Solution {
    public void helper(TreeNode root, ArrayList<Integer> ar){
        if(root==null){
            return;
        }
        helper(root.left,ar);
        helper(root.right,ar);
        ar.add(root.val);
        return;
    }
    public List<Integer> postorderTraversal(TreeNode root) {
        
        ArrayList<Integer> ar=new ArrayList<>();
        helper(root,ar);
        return ar;

    }
}
```

Binary Tree Level Order Traversal

```python
class Solution {
    public List<List<Integer>> levelOrder(TreeNode root) {
        List<List<Integer>> L=new ArrayList<>();
        helper(root,L);
        return L;
        
    }
    public static void helper(TreeNode root,List<List<Integer>> L){
        if(root==null){
            return;
        }
        Queue<TreeNode> q=new LinkedList<>();
        q.add(root);
        q.add(null);
        ArrayList<Integer> level=new ArrayList<>();
        while(!q.isEmpty()){
            TreeNode curr=q.remove();
            if(curr==null){
                L.add(level);
                if(q.isEmpty()){
                    break;
                }else{
                    level=new ArrayList<>();
                    q.add(null);
                }
            }else{
                level.add(curr.val);
                if(curr.left!=null){
                    q.add(curr.left);
                }
                if(curr.right!=null){
                    q.add(curr.right);
                }
            }
        }
    }
}
```

 Maximum Depth of Binary Tree

```python
class Solution {
    public int maxDepth(TreeNode root) {
        if(root==null){
            return 0;
        }
        int lh=maxDepth(root.left);
        int rh=maxDepth(root.right);
        int height=Math.max(lh,rh)+1;
        return height;
        
    }
}
```
